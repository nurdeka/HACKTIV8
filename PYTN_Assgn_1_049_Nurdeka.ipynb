## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

Bagaimana plotting histogram criminal category by year?

```python
df2 = criminal.pivot_table(index='year', columns='major_category', values='value', aggfunc=sum)
df2.head()
```

```python
df2.fillna(0, inplace=True)
df2.head(10)
```

```python
df2.plot(kind='hist', figsize=(10,6))

plt.show()

```

```python
count, bin_edges = np.histogram(df2, 10)

df2.plot(kind='hist',
        figsize=(10,6),
        bins=15,
        alpha=0.5,
        xticks=bin_edges)
plt.title('Histogram of Criminal Cases in London from 2008 - 2016') # add a title to the histogram
plt.ylabel('Frquency of Years') # add y-label
plt.xlabel('Range Number of Cases') # add x-label
plt.show()
```

**Insight:**

Terlihat histogram tingkat kriminal di London tahun 2008 - 2016 untuk kategori Theft and Handling mempunyai kecenderungan rentang yang cenderung kecil antara 72000 - 77000 pada kesemua tahun. \
Sedangkan kategori Violence Agains The Person cenderung mempunyai rentang yang cukup besar antara 54000 - 69000 selama 9 tahun.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Bagaimana tingkat sebaran data kriminalitas per bulannya?

```python
df.head()
```

```python
criminal['DATE'] = pd.to_datetime(criminal[['year', 'month']].assign(DAY=1))
criminal
```

```python
criminal_date1 = criminal.groupby('DATE', axis=0).sum()
criminal_date1 = criminal_date1.reset_index()
criminal_date1
```

```python
norm_value = (criminal_date1['value'] - criminal_date1['value'].min())/(criminal_date1['value'].max() - criminal_date1['value'].min())
norm_value
```

```python
criminal_date2 = criminal.pivot_table(index='DATE', columns='major_category', values='value', aggfunc=sum)
criminal_date2.fillna(0, inplace=True)
criminal_date2 = criminal_date2.reset_index()
criminal_date2.head()
```

```python
norm_burglary = (criminal_date2['Burglary'] - criminal_date2['Burglary'].min())/(criminal_date2['Burglary'].max() - criminal_date2['Burglary'].min())
norm_crim = (criminal_date2['Criminal Damage'] - criminal_date2['Criminal Damage'].min())/(criminal_date2['Criminal Damage'].max() - criminal_date2['Criminal Damage'].min())
norm_theft = (criminal_date2['Theft and Handling'] - criminal_date2['Theft and Handling'].min())/(criminal_date2['Theft and Handling'].max() - criminal_date2['Theft and Handling'].min())

print(norm_burglary)
print(norm_crim)
print(norm_theft)
```

```python
criminal_date1.plot(kind='scatter', x='DATE', y='value', figsize=(10,6), color='m', s=norm_value*500)
plt.title('Scatter Plot')
plt.xlabel('Date')
plt.ylabel('Value')

plt.show()
```

```python
#Bulgrary
ax0 = criminal_date2.plot(kind='scatter',
                         x='DATE',
                         y='Burglary',
                         figsize=(14, 8),
                         alpha=0.5,
                         color='g',
                         s=norm_burglary * 2000 +10)
#Criminal Damage
ax1 = criminal_date2.plot(kind='scatter',
                         x='DATE',
                         y='Criminal Damage',
                         alpha=0.5,
                         color='b',
                         s=norm_crim * 2000 +10,
                         ax=ax0)

#Criminal Damage
ax2 = criminal_date2.plot(kind='scatter',
                         x='DATE',
                         y='Theft and Handling',
                         alpha=0.5,
                         color='r',
                         s=norm_theft * 2000 +10,
                         ax=ax0)
ax0.set_ylabel('Number of Crime')
ax0.set_title('Criminal Cases in London')
ax0.legend(['Bulgrary', 'Criminal Damage', 'Theft'], loc='upper left', fontsize='large')
```

**Insight:**

Secara akumulasi, sebaran tingkat kriminalitas per bulannya menunjukkan tren yang mengalami peningkatan sejak tahun 2014 hingga 2016. \
Untuk tingkat kriminalitas kategorik bulanan terbanyak masih dipegang kategori Theft and Handling dengan kecenderungan meningkat di tahun 2016.

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

#!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

print ('Wordcloud is installed and imported!')
```

```python
df_word = criminal['major_category']
df_word
```

```python
stopword = set(STOPWORDS)
```

```python

wordcloud2 = WordCloud(width=1600, height=800, max_font_size=200, background_color='white',
                       max_words=2000, stopwords=stopword).generate(' '.join(df_word.values))
plt.figure(figsize=(12,10))
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
